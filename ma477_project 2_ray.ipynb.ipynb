<h2>Data Processing</h2>

```python
# importing several libraries that I think i might need 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_validate, cross_val_score, ShuffleSplit,  KFold
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.metrics import r2_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import auc, roc_curve
from sklearn.linear_model import LogisticRegression
from collections import defaultdict
```

```python
# importing the datasets
# before importing the datasets I removed the columns including indexing from the excel files. I did this because they are
# automatically indexed after importing in python and I found it easier to do in excel that in python
df_train=pd.read_csv('airline_train.csv')
df_test=pd.read_csv('airline_test.csv')
```

```python
# check the head of the data
df_train.head()
```

```python
df_test.head()
```

After importing the datasets I am going to convert the variables that have a string value with two possible responses into boolean responses with the following conventions.

<h3>Gender</h3>

 $0$ = Female
 
 $1$ = Male
 
 <h3>Customer Type</h3>

 $0$ = Disloyal Customer
 
 $1$ = Loyal Customer
 
 <h3>Type of Travel</h3>

 $0$ = Personal Travel 
 
 $1$ = Business Travel
 
 <h3>Satisfaction</h3>

 $0$ = Neutral or Dissatisfied 
 
 $1$ = Satisfied

```python
# change values to 1s and 0s
df_train['Gender']=df_train['Gender'].apply(lambda x: 1 if x=='Male' else 0)
df_train['Customer Type']=df_train['Customer Type'].apply(lambda x: 1 if x=='Loyal Customer' else 0)
df_train['Type of Travel']=df_train['Type of Travel'].apply(lambda x: 1 if x=='Business Travel' else 0)
df_train['satisfaction']=df_train['satisfaction'].apply(lambda x: 1 if x=='satisfied' else 0)

# apply the same change to the test set
df_test['Gender']=df_test['Gender'].apply(lambda x: 1 if x=='Male' else 0)
df_test['Customer Type']=df_test['Customer Type'].apply(lambda x: 1 if x=='Loyal Customer' else 0)
df_test['Type of Travel']=df_test['Type of Travel'].apply(lambda x: 1 if x=='Business Travel' else 0)
```

```python
df_train.head()
```

I am going to handled the qualitative features with more than two possible values by turning that single column variable into "n" new columns where n=the number of possible responses with boolean responses.

```python
# convert the varibales into additional columns with boolean responses
df_train=pd.get_dummies(df_train,columns=['Class'],drop_first=False)
df_test=pd.get_dummies(df_test,columns=['Class'],drop_first=False)
```

```python
df_train.head()
```

```python
#check for null values
# plt.figure(figsize=(10,6))
# sns.heatmap(df_train.isnull()==True,cbar=False)
```

From this heat map we can see that the following variables are missing several values within the training data set: 'Inflight Entertainment', 'Baggage Handling', 'Departure Delay in Minutes', and 'Arrival Delay in Minutes'

```python
# plt.figure(figsize=(10,6))
# sns.heatmap(df_test.isnull()==True,cbar=False)
```

This heatmap shows that the variable 'Arrival Delay in Minutes' is missing two values in the testing data set

```python
# for the vales in the train set that are on a likert scale, I am going ot impute the missing values as the mean of their 
# respective columns. I think this is reasonable because the data set is so large that values in each column should follow 
# a normal distribution and imputing the mean should not affet this distribution
df_train['Inflight entertainment'].fillna(value=df_train['Inflight entertainment'].mean(),inplace=True)
df_train['Baggage handling'].fillna(value=df_test['Baggage handling'].mean(),inplace=True)
df_train['Departure Delay in Minutes'].fillna(value=df_test['Departure Delay in Minutes'].mean(),inplace=True)
df_train['Arrival Delay in Minutes'].fillna(value=df_test['Arrival Delay in Minutes'].mean(),inplace=True)

# for the two missing values in the test set I am applying the same procedure
df_test['Arrival Delay in Minutes'].fillna(value=df_test['Arrival Delay in Minutes'].mean(),inplace=True)
```

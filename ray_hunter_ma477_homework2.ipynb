<h4> Find all the listings of room type <font color='red'> Shared rooms</font> in <font color='red'> Manhattan</font> with a price range of <font color='red'> \$60$-$\$75</font> and that have received over <font color='red'>100 reviews</font></h4>

```python
#Enter your answer here
abnb[(abnb['neighbourhood_group']=='Manhattan') & (abnb['room_type']=='Shared room') & (abnb['price']<75) & (abnb['price']>60) & (abnb['number_of_reviews']>100)]
```

<h4> Find the <font color='red'>minimum</font>, <font color='red'>maximum</font>, and <font color='red'>average</font> values for both <font color='red'>price</font> and <font color='red'> number of reviews</font> for the listings in each of the different neighbourhoods of Manhattan</h4>

<font color='red'>5pts</font>

```python
#Your answer
abnb_m.groupby(by='neighbourhood').agg({'price':[np.min,np.max,np.mean],'number_of_reviews':[np.min,np.max,np.mean]})
```
